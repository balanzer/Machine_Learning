# Tips Dataset


`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

### Q1] Import the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
sns.set(color_codes=True)
%matplotlib inline 

data = pd.read_csv('tips.csv',index_col=0)
data
```

```python
#shape 

print(f'shape of the data is {data.shape}')


data.head(5)
```

### Q3] Understand the distribution of 'tip' and 'total_bill' using different plots

Hint: Use distplot and boxplot

```python
# distplot 
sns.histplot(data,x='total_bill',kde=True);
```

```python
# distplot tips

sns.histplot(data,x='tip',kde=True);
```

```python
##box plot 

sns.boxplot(y='total_bill',data=data)
plt.show()
sns.boxplot(y='tip',data=data)
plt.show()
```

### Q4] Relationship between 'total_bill' and 'tip'

Implement jointplot, hex jointplot, pairplot and lmplot

```python
sns.jointplot(data['total_bill'], data['tip'],color='green')
plt.show()

sns.jointplot(data['total_bill'], data['tip'], kind = "hex",color='red')
plt.show()


sns.pairplot(data[['total_bill','tip']])
plt.show()


sns.lmplot(x = "total_bill",y = "tip", data = data)
plt.show()
```

### Q5] Make plots to check distribution of tip with different gender. 

Use striplot and boxplot

```python
sns.stripplot(data['sex'], data['tip'])
plt.show()
sns.boxplot(data['sex'], data['tip'])
plt.show()
sns.boxplot(data['sex'], data['tip'], hue=data['smoker'])
plt.show()
```

### Q6] Draw relationship between 'total_bill', 'tip' and 'sex'

use lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=data,fit_reg=False,hue='sex')
plt.show()
```

### Q7] Which day bring in higher bill amount?

use stripplot, swarmplot and boxplot

```python
sns.stripplot(x='day',y='total_bill',data=data,jitter=False)
plt.show()

sns.swarmplot(x='day',y='total_bill',data=data)
plt.show()

sns.boxplot(y='total_bill',x='day',data=data)
plt.show()
```

### Q8] Find out count of total males and females visiting each day.

use countplot

```python
sns.countplot(data['day'])
plt.show()

sns.countplot(data['day'], hue = data['sex'])
plt.show()


sns.countplot(data['day'], hue = data['smoker'])
plt.show()
```

### Q9] At what time of which day does the restaurant get higher amount of bill (consider for male and female separately)

use lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=data,fit_reg=False,hue='day',row='sex',col='time')
plt.show()
```

### Q10] Plot graphs for smoker and non-smoker (try drawing relationship with 'tip').

use stripplot and boxplot

```python
sns.stripplot(data['smoker'], data['tip'])
plt.show()
sns.boxplot(data['smoker'], data['tip'])
plt.show()
```
