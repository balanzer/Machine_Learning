```python
#Create Dataframe for sales


df7 = pd.DataFrame({"customer":[101,102,103,104],
                   "category":['A','B','C','D'],
                    "important":['Y','Y','N',None],
                    "sales":[10.00,15.21,1,2],
                   })

df8 = pd.DataFrame({"customer":[101,102,105,106],
                   "colour":['red','green','red','blue'],
                    "distance":[10,12,15.0,20],
                    "sales":[20.00,10,15.50,20.20],
                   })
print('\n\ndf7')
print()
print(df7)
print('\n\ndf8')
print()
print(df8)

```

```python
#data should stack on top of each other 
pd.concat([df7,df8],axis=0,sort=False)
```

```python
#data should stack on top of each other , with column sorting
pd.concat([df7,df8],axis=0,sort=True)
```

```python
#data should placed next to each other in columns 
pd.concat([df7,df8],axis=1,sort=False)
```

```python
#Merging Data Frames

print('\n\ndf7')
print()
print(df7)
print('\n\ndf8')
print()
print(df8)


```

```python
#Merging Data Frames

# outer is union operations on the 'On' attribute given - in this case customer id 


pd.merge(df7,df8,how='outer',on='customer')
```

```python
#inner - is and operations 

pd.merge(df7,df8,how='inner',on='customer')
```

```python
#left - includes all rows from 1st dataset, and matching from 2nd with all columns from 2nd data set those has matching customer id

pd.merge(df7,df8,how='left',on='customer')
```

```python
#right - includes all rows from 2nd dataset, and matching from 1st with all columns

pd.merge(df7,df8,how='right',on='customer')
```
