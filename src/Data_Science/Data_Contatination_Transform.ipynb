```python
import numpy as np
import pandas as pd

from pandas import Series,DataFrame
```

```python
rowindex=["row "+str(x) for x in range(1,7,1)]
colindex=["col "+str(x) for x in range(1,7,1)]

df_obj1 = DataFrame(np.arange(36).reshape(6,6))
df_obj1
```

```python
df_obj2 = DataFrame(np.arange(15).reshape(5,3))
df_obj2
```

## concat value

```python
pd.concat([df_obj1, df_obj2], axis=1) ## concat by adding new columns
```

```python
pd.concat([df_obj1, df_obj2], axis=0) ## concat by adding new rows
```

## Transfrom Data

```python
## Dropping Data by row 
df_obj1.drop([0,1])
```

```python
## Dropping Data by column 
df_obj1.drop([0,1], axis=1)
```

```python

#adding new data using join


series_obj = Series(np.arange(6))
series_obj.name='new data'
df_obj1.join(series_obj)
```

```python
## adding new data with append as row 

df_obj1.append(series_obj)
df_obj1.append(df_obj2, ignore_index=True)
```

```python
## Sorting Data by columns

df_obj1.sort_values(by=5, ascending=False)
```
